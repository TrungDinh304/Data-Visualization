```python
# Kiểm tra kiểu dữ liệu trong cột 'date'
unique_formats = df['date'].apply(type).unique()
print(f"Các kiểu dữ liệu duy nhất trong cột 'date': {unique_formats}")

# Kiểm tra số lượng giá trị duy nhất và giá trị trùng lặp
unique_dates = df['date'].nunique()
total_duplicates = df.shape[0] - unique_dates
print(f"Số lượng giá trị ngày duy nhất: {unique_dates}")
print(f"Số lượng bản ghi trùng lặp dựa trên cột 'date': {total_duplicates}")

# Chuyển đổi cột 'date' sang kiểu datetime và kiểm tra khoảng thời gian
df['date_converted'] = pd.to_datetime(df['date'], errors='coerce')
date_min = df['date_converted'].min()
date_max = df['date_converted'].max()
print(f"Khoảng thời gian trong dữ liệu: Từ {date_min} đến {date_max}")

# Đếm số lượng ngày không hợp lệ
invalid_dates = df['date_converted'].isnull().sum()
print(f"Số lượng giá trị ngày không hợp lệ: {invalid_dates}")

# Thêm cột 'year' và 'month' để phân tích theo năm và tháng
df['year'] = df['date_converted'].dt.year
df['month'] = df['date_converted'].dt.month

# Phân tích dữ liệu theo năm
yearly_counts = df['year'].value_counts().sort_index()
print("Số lượng bản ghi theo từng năm:")
print(yearly_counts)

# Phân tích dữ liệu theo tháng (không phân biệt năm)
monthly_counts = df['month'].value_counts().sort_index()
print("Số lượng bản ghi theo từng tháng (không phân biệt năm):")
print(monthly_counts)
```

```python
# Kiểm tra các giá trị bất thường trong cột 'rain'
invalid_rain = df[df['rain'] < 0]
print(f"Số giá trị bất thường trong cột 'rain': {len(invalid_rain)}")

# Loại bỏ các giá trị bất thường
df = df[df['rain'] >= 0]
```

```python
# Kiểm tra các giá trị duy nhất trong cột 'province' và 'wind_d'
print("Danh sách các tỉnh:", df['province'].unique())
print("Danh sách hướng gió:", df['wind_d'].unique())

# Chuẩn hóa (viết thường và loại bỏ khoảng trắng)
df['province'] = df['province'].str.strip().str.title()
df['wind_d'] = df['wind_d'].str.strip().str.upper()

# Kiểm tra lại sau chuẩn hóa
print("Danh sách các tỉnh sau chuẩn hóa:", df['province'].unique())
```

```python
# Hàm phát hiện và xử lý giá trị ngoại lai dựa trên khoảng hợp lý
def remove_outliers_with_context(df, column, min_val, max_val):
    # In khoảng giá trị hợp lý
    print(f"Xử lý cột: {column}")
    print(f" - Khoảng giá trị hợp lý: [{min_val}, {max_val}]")

    # Lưu lại các giá trị ngoại lai
    outliers = df[(df[column] < min_val) | (df[column] > max_val)]
    print(f" - Số lượng outliers: {len(outliers)}")
    print(f" - Giá trị outliers:\n{outliers[column].tolist()}\n")
    
    # Loại bỏ outliers và trả kết quả
    return df[(df[column] >= min_val) & (df[column] <= max_val)]

# Áp dụng xử lý cho từng cột với khoảng giá trị hợp lý
column_ranges = {
    'max': (5, 50),
    'min': (0, 40),
    'wind': (0, 150),
    'rain': (0, 500),
    'humidi': (10, 100),
    'pressure': (980, 1050)
}

# Lặp qua từng cột số và xử lý
for col, (min_val, max_val) in column_ranges.items():
    print(f"Trước xử lý, số dòng: {len(df)}")
    df = remove_outliers_with_context(df, col, min_val, max_val)
    print(f"Sau xử lý, số dòng: {len(df)}\n")

df['is_outlier'] = ((df['max'] < 5) | (df['max'] > 50)).astype(int)

# Xem lại dữ liệu sau xử lý
df
```

```python
# Kiểm tra xem hai cột có giống nhau không
are_columns_equal = (df['date'] == df['date_converted']).all()
print(f"Hai cột date và date_converted có giống nhau không? {are_columns_equal}")

# Nếu giống nhau, xóa cột thừa
if are_columns_equal:
    df = df.drop(columns=['date_converted'])
    print("Đã xóa cột date_converted vì trùng với date.")

df
```

```python
# Chuẩn hóa tên tỉnh/thành phố
def normalize_province_names(df, column):
    # Tạo từ điển chuẩn hóa tên
    name_mapping = {
        'Ha Noi': 'Ha Noi',
        'Hanoi': 'Ha Noi'
    }
    df[column] = df[column].replace(name_mapping)
    return df

# Áp dụng chuẩn hóa
df = normalize_province_names(df, 'province')

# Kiểm tra các tên tỉnh/thành phố
unique_provinces = df['province'].unique()
print(f"Các tỉnh/thành phố sau khi chuẩn hóa: {unique_provinces}")
```
